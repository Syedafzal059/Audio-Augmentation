```python
# make a list of all training files in the LibriSpeech Dataset

audio_folder_path = 'EFD_data/W/'
masked_spectogram_folder_path ="EFD_Augmented/EFD_Wheezels_FM_param_10/"
files = librosa.util.find_files(audio_folder_path, ext=None, recurse=True)
files
```

```python

file_path = files[0]

wav, sr = librosa.load(file_path, sr=None)
print(wav.shape, wav.max(), wav.min())
ipd.Audio(file_path)
```

```python
def show_spectrogram(wav):
    plotter = librosa.stft(wav, n_fft=480, hop_length=160,win_length=480, window='hamming')
    spect, phase = librosa.magphase(plotter)
    return spect
```

```python
i=0
for items in files:

    wav, sr = librosa.load(items, sr=None)
    spec =show_spectrogram(wav)

    
    
    
    # Convert to mel-spectrogram    
    mel_spectrogram = tfio.audio.melscale(
        spec, rate=16000, mels=128, fmin=0, fmax=8000)
    
   
#     plt.figure()
#     plt.imshow(tf.math.log(mel_spectrogram).numpy())

    # Convert to db scale mel-spectrogram
    dbscale_mel_spectrogram = tfio.audio.dbscale(
        mel_spectrogram, top_db=80)

#     plt.figure()
#     plt.imshow(dbscale_mel_spectrogram.numpy())

    
    # Freq masking
    freq_mask = tfio.audio.freq_mask(dbscale_mel_spectrogram, param=10)
    
#     plt.figure()
#     plt.imshow(freq_mask.numpy())
    
    
   # Time masking
    time_mask = tfio.audio.time_mask(dbscale_mel_spectrogram, param=10)
    
    masked_dbscale_mel_spectogram = freq_mask.numpy()   
    plt.imsave(f"{masked_spectogram_folder_path}{i}.png",masked_dbscale_mel_spectogram)
    i=i+1
#     plt.figure()
#     plt.imshow(time_mask.numpy())

    
    
    
    
    
    

```
